```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates

# Read the CSV file
df = pd.read_csv('BTC_full_1min.csv')

# Convert DateTime column to datetime type
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Create the figure
plt.figure(figsize=(20, 10))

# Plot price trends
plt.plot(df['DateTime'], df['Open'], label='Open', alpha=0.7)
plt.plot(df['DateTime'], df['High'], label='High', alpha=0.7)
plt.plot(df['DateTime'], df['Low'], label='Low', alpha=0.7)
plt.plot(df['DateTime'], df['Close'], label='Close', alpha=0.7)

# Set chart title and labels
plt.title('Bitcoin Price Trends', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)

# Set x-axis date format
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

# Rotate x-axis labels
plt.gcf().autofmt_xdate()

# Add legend
plt.legend(fontsize=12)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()

# Print the time range of the dataset
print(f"Dataset time range: from {df['DateTime'].min()} to {df['DateTime'].max()}")
```
